# 目標：
安裝Basemap並且可以使用它繪製地理資訊。這個教程中，我們將涉及地理繪圖module:Basemap。Basemap是Matplotlib的擴展。

# 重點：
Basemap在地圖集上的擴展應用

# 作業：
1. 更改Projection的設定繪製出地球儀圖形(經緯度任選)
* "Projection"參數為"ortho"，並以NASA寶石藍輸出圖像
* "Projection"參數為"Mill"，所得平展位面
* "Projection"參數為"Icc"，並以NASA寶石藍輸出
2. 查看resolution / 經緯度座標的繪圖精細度

```python
# 載入套件
from mpl_toolkits.basemap import Basemap
import matplotlib.pyplot as plt
import numpy as np
```

```python
# 創建基本地圖

#設定圖形大小
plt.figure(figsize=(5,5))

#設定投影圖形
m = Basemap(projection = 'ortho', resolution = None, lat_0 = 50, lon_0 = 0)

#設定圖形顏色，選用 NASA寶石藍輸出
m.bluemarble(scale = 0.5)

#秀出圖形
plt.show()
```

```python
m = Basemap(projection = 'mill',
            llcrnrlat = -90,
            llcrnrlon = -180,
            urcrnrlat = 90,
            urcrnrlon = 180,)
#創建邊線
m.drawcoastlines()
#畫出國家
m.drawcountries(linewidth = 2)
#畫出州界
m.drawstates(color = 'b')
#畫出城市
m.drawcounties(color = 'darkred')

plt.title('Basemap Tutorial')

plt.show()
```

```python
#set resolution = None跳過邊界數據集的處理
# projection='lcc'

m = Basemap(width=12000000,height=9000000,projection='lcc',
            resolution=None,lat_1=45.,lat_2=55,lat_0=50,lon_0=-107.)

#選用 NASA 寶石藍輸出 
m.bluemarble()

#秀圖
plt.show()
```
