###  Q1: é€éæ•¸å€¼æ³•è¨ˆç®—  Age å’Œ  Survived  æ˜¯å¦æœ‰ç›¸é—œæ€§?


```python
# Ageç‚ºé€£çºŒå‹ï¼ŒSurvivedç‚ºé›¢æ•£å‹ï¼Œæ‰€ä»¥ä½¿ç”¨ Eta Squared
```

```python
# è¨ˆç®—ç›¸é—œä¿‚æ•¸ä¸å…è¨±æœ‰éºå¤±å€¼ï¼Œæ‰€ä»¥å¿…é ˆå…ˆè£œå€¼æˆ–å°‡éºå¤±å€¼åˆªé™¤
# Step 1:å–å‡º Ageå’Œ Survived_cateè³‡æ–™
complete_data = df_train[['Age', 'Survived_cate']].dropna()
display(complete_data)
```

```python
# Step 2:é‹ç”¨ pg.anova è¨ˆç®—ä¸‰ç¨®è®Šç•°æ•¸
aov = pg.anova(dv='Age', between='Survived_cate', data=complete_data, detailed=True)
aov
```

```python
# Step 3:è®Šç•°æ•¸æ›ç®—å¾—åˆ° Eta Squared (ğœ¼^ğŸ)
etaSq = aov.SS[0]/(aov.SS[0]+aov.SS[1])
etaSq
```

```python
def judgment_etaSq(etaSq):
    if etaSq < .01:
        qual = 'Negligible'
    elif etaSq < .06:
        qual = 'Small'
    elif etaSq < .14:
        qual = 'Medium'
    else:
        qual = 'Large'
    return(qual)
judgment_etaSq(etaSq)

# çµè«–:å¹´ç´€å’Œå­˜æ´»èˆ‡å¦æ²’æœ‰ç›¸é—œ
```

```python
g = sns.catplot(x="Survived_cate", y="Age", hue="Survived_cate",
               data=complete_data, kind="violin")
```

### Q2:é€éæ•¸å€¼æ³•è¨ˆç®—  Sex å’Œ Survived  æ˜¯å¦æœ‰ç›¸é—œæ€§?


```python
# Sexç‚ºé›¢æ•£å‹ï¼ŒSurvived_cateç‚ºé›¢æ•£å‹ï¼Œæ•…è€Œä½¿ç”¨ CramÃ©r's V
# Step 1:ç”¨äº¤å‰åˆ—é€£è¡¨(contingency table)ï¼Œä¾†æ•´ç†å…©å€‹é¡åˆ¥å‹çš„è³‡æ–™
complete_data = df_train[['Sex','Survived_cate']].dropna()
complete_data
contTable = pd.crosstab(complete_data['Sex'], complete_data['Survived_cate'])
contTable
```

```python
# Step 2:è¨ˆç®—è‡ªç”±åº¦ df*
df = min(contTable.shape[0], contTable.shape[1])-1
df
```

```python
# Step 3:é‹ç”¨ researchpy å¥—ä»¶ï¼Œè¨ˆç®—å‡º Cramerâ€™s V ä¿‚æ•¸
crosstab, res = researchpy.crosstab(complete_data['Sex'], complete_data['Survived_cate'], test='chi-square')
print(res)
print('-----------------------------------------')
print("Cramer's value is",res.loc[2,'results'])
```

```python
# å¯«ä¸€å€‹å‰¯ç¨‹å¼åˆ¤æ–·ç›¸é—œæ€§çš„å¼·åº¦
def judgment_CramerV(df,V):
    if df == 1:
        if V < 0.10:
            qual = 'negligible'
        elif V < 0.30:
            qual = 'small'
        elif V < 0.50:
            qual = 'medium'
        else:
            qual = 'large'
    elif df == 2:
        if V < 0.07:
            qual = 'negligible'
        elif V < 0.21:
            qual = 'small'
        elif V < 0.35:
            qual = 'medium'
        else:
            qual = 'large'
    elif df == 3:
        if V < 0.06:
            qual = 'negligible'
        elif V < 0.17:
            qual = 'small'
        elif V < 0.29:
            qual = 'medium'
        else:
            qual = 'large'
    elif df == 4:
        if V < 0.05:
            qual = 'negligible'
        elif V < 0.15:
            qual = 'small'
        elif V < 0.25:
            qual = 'medium'
        else:
            qual = 'large'
    else:
        if V < 0.05:
            qual = 'negligible'
        elif V < 0.13:
            qual = 'small'
        elif V < 0.22:
            qual = 'medium'
        else:
            qual = 'large'
    return(qual)

judgment_CramerV(df,res.loc[2,'results'])
```
